```python
import Functions as f
import pandas as pd
```

```python
data = pd.read_csv('data/diabetes.csv')
# reduce samples to 600
data = data.sample(600)
X = data.drop('Outcome', axis=1).to_numpy()
y = data['Outcome'].to_numpy()
```

```python
iterations = 100
error = 0.0001
alpha = 0.001
```

- My implementation of Sammon projection is in Functions.py as I wanted to reuse it for task 3 and to have a better notebook structure.

```python
y_sammon = f.sammon(X, iterations, error, alpha)
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(1, figsize=(20, 10))
ax.scatter(y_sammon[:, 0], y_sammon[:, 1], c=y)
ax.set_title('Sammon projection of diabetes data')
```

- After many test runs, I found that the best alpha is 0.001. Higher alpha always produces points exploding (points very far away from the rest of the data). I also tried testing with different iterations and error but it doesn't seem to affect the result much.
