```python
iterations = 100
error = 0.0001
alpha = 0.001
```

```python
y_sammon = f.sammon(X, iterations, error, alpha)
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(1, figsize=(20, 10))
ax.scatter(y_sammon[:, 0], y_sammon[:, 1], c=y)
ax.set_title('Sammon projection without normalization')
```
