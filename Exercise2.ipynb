```python
import numpy as np
import Functions as f
import pandas as pd
```

```python
data = pd.read_csv('data/diabetes.csv')
X = data.drop('Outcome', axis=1).to_numpy()
y = data['Outcome'].to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

```python
iterations = 100
error = 0.0001
alpha = 0.3
```

```python
y_sammon = f.sammon(X, iterations, error, alpha)
```

```python
y_sammon_normalized = f.sammon(X_normalized, iterations, error, alpha)
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].scatter(y_sammon[:, 0], y_sammon[:, 1], c=y)
ax[0].set_title('Sammon projection without normalization')
ax[1].set_title('Sammon projection with normalization')
ax[1].scatter(y_sammon_normalized[:, 0], y_sammon_normalized[:, 1], c=y)
```
